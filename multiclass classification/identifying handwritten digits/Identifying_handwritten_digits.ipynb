```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

#predefined dataset. Check documentation for more info
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

The next code converts the above binary array to an image so that view can the image and not the binary code of the image

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

So in the next cell we are getting the answer of the above images. Like 0 corresponds to the 1st image, 1 to the 2nd and so on..

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[71])
```

```python
digits.target[71]
```

```python
model.predict([digits.data[71]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted=model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
