```python
import ast
```

```python
# Converting the HTML into a normal text in the list

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
movies
```

```python
def convert1(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
    return L
```

```python
movies['cast']=movies['cast'].apply(convert1)
movies
```

```python
movies.iloc[0].crew
```

```python
# Extracting Director name 

def convert2(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(convert2)
movies
```

```python
movies=movies.dropna()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
movies['overview']=movies['overview'].apply(lambda x: [i.replace(" ","") for i in x])
movies
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies
```

```python
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['overview']
movies
```

```python
working_df=movies[['title','tags']]
working_df
```

```python
working_df.tags[0]
```

```python
def filter(tags):
    words_to_remove = ['in', 'and', 'or', 'an', 'the', 'but', 'on','a','to','is','be','has','had','as','it','of','him','from']
    return [tag for tag in tags if tag.lower() not in words_to_remove]

working_df['tags'] = working_df['tags'].apply(filter)
```

```python
working_df.tags[0]
```

```python
working_df
```

```python
# Converting it into a single string

working_df['tags']=working_df['tags'].apply(lambda x:' '.join(x))
working_df
```

```python
working_df = working_df.merge(movies[['movie_id']], left_index=True, right_index=True)
```

```python
working_df
```

```python
working_df['tags']=working_df['tags'].apply(lambda x: x.lower())
```

```python
working_df['tags'][0]
```

```python
# Feature Extraction using CountVectoriser
## CountVectorizer convert the given data into vectors based on their frequency
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(working_df['tags']).toarray().shape
```

```python
# Converting Vector to array

vectors=cv.fit_transform(working_df['tags']).toarray()
```

```python
vectors[0]
```

```python
# nltk is used for building python programs that working in natural language. 
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Stemming-reducing words to its base or root form

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
working_df['tags']=working_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Cosine similarity measures the difference between the two vectors
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    try:
        movie_index = working_df[working_df['title']==movie].index[0]
        distances = similarity[movie_index]
        movies_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

        for i in movies_list:
            print(working_df.iloc[i[0]].title)
    except:
        print("Movie Not Found")
```

```python
recommend('Avatar')
```
