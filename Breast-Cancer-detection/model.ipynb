```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df.drop(['id'], axis=1, inplace=True)
df.head(5)
```

```python
sns.countplot(x=df.diagnosis)
```

```python
df[df['diagnosis']=='M'].shape
```

```python
df[df['diagnosis']=='B'].shape
```

```python
b_df = df.loc[df['diagnosis']=='B'][:212]
b_df.head(3)
```

```python
m_df = df[df['diagnosis']=='M']
```

```python
df = pd.concat([b_df, m_df])
```

```python
sns.countplot(x=df.diagnosis)
```

```python
print("The unique values in the datframe are", df.diagnosis.unique())
```

```python
df.diagnosis = df.diagnosis.apply(lambda x:1 if x=='M' else 0)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df = pd.DataFrame(scaler.fit_transform(df), columns=b_df.columns)
df
```

```python
target = df.diagnosis
df.drop(['diagnosis'], axis =1, inplace=True)
df.head(3)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
model = Sequential([
    Dense(units = 500, activation='relu'),
    Dense(units = 350, activation = 'relu'),
    Dense(units = 100, activation='relu'),
    Dense(units = 50, activation='relu'),
    Dense(units = 25, activation='relu'),
    Dense(units = 1, activation='sigmoid')
])
model.compile(
    loss='binary_crossentropy', 
    optimizer='adam',
    metrics=['accuracy'])

X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.2)
model.fit(X_train, y_train, epochs=100, batch_size=2)
```

```python
scores = model.evaluate(X_test, y_test)
print(f"accuracy is {scores[1]*100:.2f}%")
```
